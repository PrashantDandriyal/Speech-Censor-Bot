````
def preprocess(org_aud_path):
  tx = wave.open(org_aud_path, 'r')
  print ("parameters:",tx.getparams())
  if(tx.getnchannels() > 1):
    #Convert stereo to mono
    print("Converting audio from stereo to mono")
    !sox "$org_aud_path" processed.wav channels 1

    print("Processed file into the same path with name 'processed.wav' ")
    #Remove old file
    !rm -r "$org_aud_path"
    #Rename new file to old file's name
    !mv "processed.wav" "$org_aud_path"

preprocess(wav_path)
````

````
#Finally installing aeneas
!sudo pip install aeneas[full] 
from aeneas.executetask import ExecuteTask
from aeneas.task import Task
````

````
#Create task objects that hold relevant configurations
# create Task object
config_string = "task_language=eng|is_text_type=plain|os_task_file_format=json"
task = Task(config_string=config_string)
task.audio_file_path_absolute = wav_path
task.text_file_path_absolute = text_path
task.sync_map_file_path_absolute = "/content/syncmap.json"

#Run the process task
ExecuteTask(task).execute()
# output sync map to file
task.output_sync_map_file()
````
